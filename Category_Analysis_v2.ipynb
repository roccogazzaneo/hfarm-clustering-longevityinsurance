# Product categorization - 3 step approach

## 1. Import the detailed categorization for the 2000 most frequent products and apply the new categories to the dataset

```python
from dictionary_new import category_product_dictionary
```

```python
def create_new_category(category_name, list_of_products, df):
    df[category_name] = df['desc'].apply(lambda x: 1 if x in list_of_products else 0)
    
for category in category_product_dictionary:
    create_new_category(category, category_product_dictionary[category], df)
```

### How many products are categorized after step 1?

```python
currently_categorized = df[(df[list(category_product_dictionary.keys())].T != 0).any()]
print(f"Products currently categorized: {len(currently_categorized)}")
print(f"Total number of products: {len(df)}")
print(f"In percentage: {round(len(currently_categorized) * 100 / len(df), 0)} %")
```

```python
non_categorized_1 = df[(df[list(category_product_dictionary.keys())].T == 0).all()]
```

## 2. Categorize some products using external keywords

```python
keywords = {'Contains plastic': ['filmado'], 'Fruit' : ['aguacate', 'albaricoque', 'arandano', 'cereza', 'ciruela', 'coco', 
                                            'frambuesa', 'fresa', 'granada', 'higo', 'kiwi', 'lima', 'limon', 'mango',
                                            'manzana', 'maracuya', 'melocoton', 'membrillo', 'mora', 'naranja', 'papaya',
                                            'pera', 'piña', 'platano', 'sandia', 'uva'],
'Vegetable': ['alcachofa', 'apio', 'batata', 'berenjena', 'brocoli', 'calabacin', 'cebolla', 'coliflor', 'esparrago', 
              'guisante', 'judia', 'lechuga', 'papa', 'pepino', 'rabano', 'remolcha', 'tomate', 'zamallo', 'fresa', 'freson'],
'Dairy':['yogur', 'leche'],
'Junk':['chocolate', 'pizza', 'frito'],
'Ready to eat/drink' : ['cocido'],
'High in sugar' : ['helado', 'miel']}
```

```python
def update_category_keywords(category_name, list_of_products, df):
    df[category_name] = df['desc'].apply(lambda x: int(any(keyword in x for keyword in list_of_products)))
    
for category in keywords:
    update_category_keywords(category, keywords[category], non_categorized_1)
```

### How many products are categorized after step 2?

```python
df = pd.concat([currently_categorized, non_categorized_1], axis = 0)
step_2 = df[(df[list(category_product_dictionary.keys())].T != 0).any()]
print(f"Products currently categorized: {len(step_2)}")
print(f"Total number of products: {len(df)}")
print(f"In percentage: {round(len(step_2) / len(df) * 100, 0)} %")
```

## 3. Categorize the remaining products by looking at the most frequent items already categorized

```python
non_categorized = df[(df[list(category_product_dictionary.keys())].T == 0).all()]
```

```python
import re
frequency_per_category = dict()
for category in category_product_dictionary:
    most_frequent_words = dict()
    for string in category_product_dictionary[category]:
        words = re.findall(r'\w+', string)
        for word in words:
            if word in most_frequent_words:
                most_frequent_words[word] += 1
            else:
                most_frequent_words[word] = 1
    frequency_per_category[category] = most_frequent_words
```

### Sort per category

```python
frequency_per_category
```

```python
for category in frequency_per_category:
    frequency_per_category[category] = {k: v for k, v in sorted(frequency_per_category[category].items(), key=lambda item: item[1], reverse=True)}
```

### Take only sensible words that appear frequently

```python
threshold_per_category = {'Carrefour product': 8, 
                          'Contains plastic': 4, 
                          'Very Salty':3, 
                          'Junk':4, 
                          'Raffinato':5, 
                          'Fruit':3, 
                          'Vegetable':3, 
                          'Family product':3, 
                          'Alcohol':3, 
                          'Ready to eat/drink': 3, 
                          'High in sugar': 5, 
                          'Packed in glass or aluminum':3, 
                          'Condimenti':4, 'Integrale':4, 
                          'Egg': 3, 'Fish':5, 
                          'White meat':19, 
                          'Red meat': 3, 'Dairy':3, 'Drinks':3, 'Coffee': 3, 'For pets':1, 'Legumes': 1, 
                          'Fitness Products':1, 'Cereals':1, 'Instruments to cook':3, 'One-time use product':3, 
                          'House products': 3, 'Hygiene':3, 'Cosmetics':1, 'For children':3, 'Medical':1,
                          'Discounts':1, 'Snacks':1, 'Cook':1, 'Drinks':1}

for category in frequency_per_category:
    try:
        frequency_per_category[category] = {k: v for k, v in list(frequency_per_category[category].items()) if v >= threshold_per_category[category]}
    except KeyError:
        print(category)
```

```python
frequency_per_category['Drinks'] = ["limonada", "jugo", "te", "rojita", "fanta", "pepsi", "coca"]
frequency_per_category['Clothes'] = ["pantalones", "bufanda", "falda", "gorro", "camiseta", "camisa"]
frequency_per_category['Woman'] = ["femenina", "woman", "mujer", "colonia", "maquillaje", "lapiz", "labial"]
frequency_per_category['Man'] = ["cinturon", "fusion", "gillet", "maquinilla"]
frequency_per_category['Plants'] = ["rosa", "tulipan", "cedro", "castaño"]
```

```python
for category in frequency_per_category:
    try:
        frequency_per_category[category] = list(frequency_per_category[category].keys())
    except AttributeError:
        frequency_per_category[category] = frequency_per_category[category]
```

### Categorize

```python
for category in frequency_per_category:
    non_categorized[category] = non_categorized['desc'].apply(lambda x: 1 if any(i in frequency_per_category[category] for i in re.findall(r'\w+', x)) else 0)
```
